# Neural NetWork for Fashion_MNIST

```python
import tensorflow as tf
from tensorflow import keras
print(tf.__version__)
```

# Bringing The Dataset

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(train_images[0])
```

#### the image has Value from 0 ~ 255 but neural networks work better with normalize data so it's better to change the value to be between zero and one by dividing them by 255

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

# Building the Model
#### 3 Layers with hidden Layer

```python
model = tf.keras.Sequential([tf.keras.layers.Flatten(),
                            tf.keras.layers.Dense(128, activation = tf.nn.relu),
                            tf.keras.layers.Dense(10, activation = tf.nn.softmax)
])
```

# the hidden layer is to figureout the rule

```python
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy')

model.fit(train_images, train_labels, epochs = 5)
```

```python
model.evaluate(test_images, test_labels)
```
